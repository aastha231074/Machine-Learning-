```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Position_Salaries.csv')
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures 
ploy_reg = PolynomialFeatures(degree=4)
X_ploy = ploy_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_ploy, y)
```

```python
plt.scatter(X,y, color = 'red', label = 'data points')
plt.plot(X, lin_reg.predict(X), color = 'blue', label = 'Regression Line')
plt.title('Salary Vs Position (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(X,y, color = 'red', label = 'data points')
plt.plot(X, lin_reg_2.predict(ploy_reg.fit_transform(X)), color = 'blue', label = 'Regression Line')
plt.title('Salary Vs Position (Polynomial Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
