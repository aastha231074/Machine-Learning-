```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
regressor = LinearRegression()
regressor = regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = 'red', label = 'dataset')
plt.plot(X_train, regressor.predict(X_train), color = 'blue', label = 'regression line')
plt.title('Training Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'red', label = 'dataset')
plt.plot(X_test, y_pred, color = 'blue', label = 'regression line')
plt.title('Testing Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# Making a single prediction (for example salary of an employee with 12 years of experience)
# The "predict" method always excepts a 2D array as a format of inputs 
print(regressor.predict([[12]]))
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
